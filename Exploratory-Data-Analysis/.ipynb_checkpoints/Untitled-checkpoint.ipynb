```python
OUT_FOLDER = '../Output-data/'
IN_FILE = '../Data/test.txt'
```

## Define positions of data elements as described in documentation

There are two types of files: status and dynamic

```python
# Define positions for dynamic files

start_dyn = [0,9,32,36,38,46,52,54,56,62,71,75,76,82,84]
end   = [9,32,36,38,46,52,54,56,62,71,75,76,82,84,85]

# index of data elements to access lists 'start' and 'end'
ID_DYN      = 0
NAME_DYN    = 1
AGENCY_DYN  = 2
ACCES_IND_DYN=3
EFFEC_DT_DYN= 4
AGE_DYN     = 5
PAY_DYN     = 6
GRADE_DYN   = 7
LOS_DYN     = 8
STATION_DYN = 9
OCCUP       =10
OCCUP_CAT   =11
ADJ_PAY     =12
TYPE_APP    =13
WORK_SCHE   =14
```

```python
_id=[]
_name=[]
_agency=[]
_accession=[]
_effec_dt=[]
_age=[]
_pay=[]
_grade=[]
_los=[]
_duty=[]
_occup=[]
_occup_cat=[]
_adj_pay=[]
_typ_app=[]
_work_sche=[]

f = open(IN_FILE, 'r')

for line in f:
    
    _id.append (line[start[ID_DYN]:end[ID_DYN]])
    _name.append (line[start[NAME_DYN]:end[NAME_DYN]])
    _agency.append (line[start[AGENCY_DYN]:end[AGENCY_DYN]])
    _accession.append (line[start[ACCES_IND_DYN]:end[ACCES_IND_DYN]])
    _effec_dt.append (line[start[EFFEC_DT_DYN]:end[EFFEC_DT_DYN]])
    _age.append (line[start[AGE_DYN]:end[AGE_DYN]])
    _pay.append (line[start[PAY_DYN]:end[PAY_DYN]])
    _grade.append (line[start[GRADE_DYN]:end[GRADE_DYN]])
    _los.append (line[start[LOS_DYN]:end[LOS_DYN]])
    _duty.append (line[start[STATION_DYN]:end[STATION_DYN]])
    _occup.append (line[start[OCCUP]:end[OCCUP]])
    _occup_cat.append (line[start[OCCUP_CAT]:end[OCCUP_CAT]])
    _adj_pay.append (line[start[ADJ_PAY]:end[ADJ_PAY]])
    _typ_app.append (line[start[TYPE_APP]:end[TYPE_APP]])
    _work_sche.append (line[start[WORK_SCHE]:end[WORK_SCHE]])

f.close()  
```

```python
dod_dyn = pd.DataFrame({'empid': _id, 'agency': _agency, 'accession_DYN': _accession, 'effective_dt': _effec_dt,
                               'age': _age, 'pay_plan': _pay, 'grade': _grade, 'los': _los, 'duty_station': _duty, 'occupation': _occup ,
                                'occupation_cat':_occup_cat, 'adjusted_pay':_adj_pay, 'type_appointment': _typ_app, 'work_schedule': _work_sche })
```
